#### 3.4 Limitações

&nbsp; &nbsp; &nbsp; Existem algumas limitações para esses dados, então temos que ter em mente que podemos encontrar algum viés de amostragem. Esses são alguns pontos que vale a pena ressaltar:

- **Tamanho da amostra pequeno:** Trinta usuários não é um tamanho de amostra ideal, especialmente quando os dados de saúde e estilo de vida variam em diferentes facetas da sociedade. Não temos certeza se a amostra é representativa da população como um todo.
- **Dados demográficos ausentes:** dados demográficos importantes, como sexo e idade, não foram identificados. Esta é uma informação crucial que falta, pois os insights obtidos podem não refletir as diferenças na fisiologia e nos padrões de atividade entre diferentes grupos.
- **Período de coleta de dados:** o tempo de pesquisa dos dados (31 dias) não favorece para ter uma noção mais ampla das tendências sazonais, já que os hábitos de exercício do usuário diferem entre o verão e o inverno. 
- **Falta de alguns metadados:** informações como localização, estilo de vida, clima, temperatura, umidade, etc. forneceriam um contexto mais profundo aos dados obtidos.

Vamos utilizar três funções:
* **dim()**: é usada para determinar as dimensões de um objeto, como uma matriz, dataset ou array. (Linhas x Colunas)
* **head()**: é usada para visualizar as primeiras linhas de um conjunto de dados. Por padrão ela exibe as primeiras 6 linhas, mas pode-se especificar outra numeração.
* **unique()**: é usada para extrair os valores exclusivos de um vetor ou coluna de um dataframe.

````r
# Função auxiliar 
print_dataset_info <- function(dataset_name, dataset) {
    separator <- rep("-", 42)
    cat("\n\n", separator, "\n")
    cat("\n\n", dataset_name, ": ", dim(dataset)[1], "x", dim(dataset)[2], "\n")
    cat(" Usuários:", length(unique(dataset$Id)), "\n\n")
    head(dataset)
}

# Imprime as informações de cada Dataset
print_dataset_info("Activity", activity)
print_dataset_info("Calories", calories)
print_dataset_info("Intensities", intensities)
print_dataset_info("Steps", steps)
print_dataset_info("Sleep", sleep)
print_dataset_info("Heart rate", heart_rate)
print_dataset_info("Weight", weight)
````

<a id="valores-null"></a>
#### 4.4.1 Checando valores ausentes
&nbsp; &nbsp; &nbsp; Na verificação de valores ausentes, examinaremos cada um dos conjuntos de dados para identificar e tratar possíveis lacunas em nossos dados. Caso seja necessário, realizaremos a remoção ou aplicação de técnicas de preenchimento para corrigir esses espaços em branco.

````r
# Atribuindo os datasets em uma lista
datasets_list <- list(
  Activity = activity,
  Calories = calories,
  Intensities = intensities,
  Steps = steps,
  Sleep = sleep,
  Heart_rate = heart_rate,
  Weight = weight
)

# Função para checar valores nulos em um dataset
check_missing_data <- function(dataset) {
  missing_data <- is.na(dataset)
  num_missing_per_column <- colSums(missing_data)
  return(num_missing_per_column)
}

# Itera em um loop para checar todos os datasets da lista
for (dataset_name in names(datasets_list)) {
  missing_values <- check_missing_data(datasets_list[[dataset_name]])
  cat("Dataset:", dataset_name, "\n\n")
  print(missing_values)
  separator <- rep("-", 42)
  cat("\n", separator, "\n\n")
}
````

&nbsp; &nbsp; &nbsp; Com esse resultado, é possível notar que não há valores ausentes em nossos Datasets, com exceção de um: o atributo "*Fat*" da tabela **Weight** com 65 valores. Portanto, iremos remover essa coluna para garantir a integridade dos dados utilizados em nossa análise.

````r
weight <- weight %>% select(-Fat)
````

<a id="duplicatas"></a>
#### 4.4.2 Checando duplicatas
&nbsp; &nbsp; &nbsp; É fundamental conferir se há duplicação de registros em nossos dados. Caso identifiquemos duplicatas, tomaremos medidas para manter apenas uma ocorrência e remover as demais, garantindo a integridade dos nossos conjuntos de dados. Além disso, eliminaremos a coluna "TrackerDistance" do dataset **activity**, visto que durante a prévia dos dados foi possível notar que ela contém informações duplicadas de "TotalDistance".

````r
# Remove a coluna TrackerDistance
activity <- activity %>% select(-TrackerDistance)

# Função para checar e tratar duplicatas
check_and_treat_duplicates <- function(dataset_name, dataset) {
  duplicate_rows <- duplicated(dataset, fromLast = TRUE)
  if (any(duplicate_rows)) {
    duplicated_rows_count <- sum(duplicate_rows)
    cat(paste("O dataset", dataset_name, "possui valores duplicados em", duplicated_rows_count, "linhas.\n"))
    treated_dataset <- dataset[!duplicate_rows, ]
    cat(paste("Foi mantida apenas uma ocorrência das linhas duplicadas no dataset", dataset_name, ".\n"))
    return(treated_dataset)
  } else {
    return(dataset)
  }
}

# Para cada dataset
activity <- check_and_treat_duplicates("Activity", activity)
calories <- check_and_treat_duplicates("Calories", calories)
intensities <- check_and_treat_duplicates("Intensities", intensities)
steps <- check_and_treat_duplicates("Steps", steps)
sleep <- check_and_treat_duplicates("Sleep", sleep)
heart_rate <- check_and_treat_duplicates("Heart_rate", heart_rate)
weight <- check_and_treat_duplicates("Weight", weight)
````

&nbsp; &nbsp; &nbsp; Agora é necessário abordar o formato das horas. Nossos conjuntos de dados utilizam o tipo 'char' para representar dias e horas, então faremos a conversão necessária. Além disso, a separação dos dados em componentes individuais pode ser extremamente útil na etapa de análise. Vejamos o que será realizado:

* Separação dos **dias** e das **horas**

Antes de prosseguirmos, vamos renomear algumas colunas para evitar possíveis problemas no futuro.

````r
heart_rate <- heart_rate %>%
  rename(TimeofDay = Time)

weight <- weight %>%
  rename(Day = Date)

activity <- activity %>%
  rename(Date = ActivityDate)
````

&nbsp; &nbsp; &nbsp; Em todos os conjuntos de dados vamos separar a data e a hora em colunas distintas, exceto no conjunto de dados **activity**, uma vez que este contém somente informações sobre os dias.

````r
# Separa Data e Hora para cada dataset
calories$Date <- as.Date(calories$ActivityHour, format = "%m/%d/%Y %I:%M:%S %p")
calories$Time <- format(as.POSIXct(calories$ActivityHour, format = "%m/%d/%Y %I:%M:%S %p"), format = "%I:%M:%S %p")

intensities$Date <- as.Date(intensities$ActivityHour, format = "%m/%d/%Y %I:%M:%S %p")
intensities$Time <- format(as.POSIXct(intensities$ActivityHour, format = "%m/%d/%Y %I:%M:%S %p"), format = "%I:%M:%S %p")

steps$Date <- as.Date(steps$ActivityHour, format = "%m/%d/%Y %I:%M:%S %p")
steps$Time <- format(as.POSIXct(steps$ActivityHour, format = "%m/%d/%Y %I:%M:%S %p"), format = "%I:%M:%S %p")

sleep$Date <- as.Date(sleep$SleepDay, format = "%m/%d/%Y")
sleep$Time <- format(as.POSIXct(sleep$SleepDay, format = "%m/%d/%Y %I:%M:%S %p"), format = "%I:%M:%S %p")

heart_rate$Date <- as.Date(heart_rate$TimeofDay, format = "%m/%d/%Y %I:%M:%S %p")
heart_rate$Time <- format(as.POSIXct(heart_rate$TimeofDay, format = "%m/%d/%Y %I:%M:%S %p"), format = "%I:%M:%S %p")

weight$Date <- as.Date(weight$Day, format = "%m/%d/%Y")
weight$Time <- format(as.POSIXct(weight$Day, format = "%m/%d/%Y %I:%M:%S %p"), format = "%I:%M:%S %p")
````

&nbsp; &nbsp; &nbsp; Tendo criado duas novas colunas para nossos dados, podemos agora remover a coluna anterior para evitar informações redundantes. Em seguida, podemos tratar o tipo de formatação dessas novas colunas.

````r
calories <- calories %>%
  select(-ActivityHour)

intensities <- intensities %>%
  select(-ActivityHour)

steps <- steps %>%
  select(-ActivityHour)

sleep <- sleep %>%
  select(-SleepDay)

heart_rate <- heart_rate %>%
  select(-TimeofDay)

weight <- weight %>%
  select(-Day)
````

[↑](#indice)

&nbsp; &nbsp; &nbsp; Nesta etapa algumas funções serão necessárias, mas iremos apresentar as duas que fazem realmente o trabalho:

* **mdy()**: é usado para analisar e criar datas a partir de cadeias de caracteres no formato *mês/dia/ano*.
* **as_hms()**: é usado para converter uma duração de tempo (formatado como ```hh:mm:ss```) em um objeto de período que representa horas, minutos e segundos.

````r
# Converte o formato das colunas para date-time
activity$Date <- mdy(activity$Date)
calories$Time <- as_hms(format(strptime(calories$Time, format = "%I:%M:%S %p"), format = "%H:%M:%S"))
intensities$Time <- as_hms(format(strptime(intensities$Time, format = "%I:%M:%S %p"), format = "%H:%M:%S"))
steps$Time <- as_hms(format(strptime(steps$Time, format = "%I:%M:%S %p"), format = "%H:%M:%S"))
sleep$Time <- as_hms(format(strptime(sleep$Time, format = "%I:%M:%S %p"), format = "%H:%M:%S"))
heart_rate$Time <- as_hms(format(strptime(heart_rate$Time, format = "%I:%M:%S %p"), format = "%H:%M:%S"))
weight$Time <- as_hms(format(strptime(weight$Time, format = "%I:%M:%S %p"), format = "%H:%M:%S"))
````

&nbsp; &nbsp; &nbsp; Agora as colunas estão nos formatos apropriados! Além disso, os horários foram convertidos para o formato de 24 horas, o que pode proporcionar uma visualização mais clara no futuro. Podemos então finalmente adicionar mais uma coluna que irá representar os dias da semana, utilizaremos a função:

* **weekdays()**: é usado para extrair os nomes dos dias da semana de um objeto Date ou POSIXct.

````r
activity$DayOfWeek <- weekdays(activity$Date)
calories$DayOfWeek <- weekdays(calories$Date)
intensities$DayOfWeek <- weekdays(intensities$Date)
steps$DayOfWeek <- weekdays(steps$Date)
sleep$DayOfWeek <- weekdays(sleep$Date)
heart_rate$DayOfWeek <- weekdays(heart_rate$Date)
weight$DayOfWeek <- weekdays(weight$Date)
````

&nbsp; &nbsp; &nbsp; Por fim, como todo o nosso estudo de caso está sendo conduzido em português, é apropriado converter os nomes dos dias da semana para garantir maior facilidade e alinhamento.

````r
# Função para substituir os nomes da semana para Português
replace_day_names <- function(dataset) {
  dataset$DayOfWeek[dataset$DayOfWeek == "Sunday"] <- "Domingo"
  dataset$DayOfWeek[dataset$DayOfWeek == "Monday"] <- "Segunda-feira"
  dataset$DayOfWeek[dataset$DayOfWeek == "Tuesday"] <- "Terça-feira"
  dataset$DayOfWeek[dataset$DayOfWeek == "Wednesday"] <- "Quarta-feira"
  dataset$DayOfWeek[dataset$DayOfWeek == "Thursday"] <- "Quinta-feira"
  dataset$DayOfWeek[dataset$DayOfWeek == "Friday"] <- "Sexta-feira"
  dataset$DayOfWeek[dataset$DayOfWeek == "Saturday"] <- "Sábado"
  
  return(dataset)
}

# Para cada dataset
activity <- replace_day_names(activity)
calories <- replace_day_names(calories)
intensities <- replace_day_names(intensities)
steps <- replace_day_names(steps)
sleep <- replace_day_names(sleep)
heart_rate <- replace_day_names(heart_rate)
weight <- replace_day_names(weight)
````

#### 5.3 Níveis de atividade

[↑](#indice)

&nbsp; &nbsp; &nbsp; Em nosso conjunto de dados de atividade, encontram-se informações sobre os níveis de atividade dos usuários registradas por minuto. Esses dados podem ser aproveitados para uma análise abrangente dos níveis gerais de atividade de todos os usuários, usando a porcentagem como métrica.

````r
# Calcula minutos totais
activity$TotalMinutes <- activity$SedentaryMinutes +
                          activity$LightlyActiveMinutes +
                          activity$FairlyActiveMinutes +
                          activity$VeryActiveMinutes

# Calcula minutos totais para cada atividade
total_sedentary_minutes <- sum(activity$SedentaryMinutes)
total_lightly_active_minutes <- sum(activity$LightlyActiveMinutes)
total_fairly_active_minutes <- sum(activity$FairlyActiveMinutes)
total_very_active_minutes <- sum(activity$VeryActiveMinutes)

# Porcentagem
sedentary_percentage <- (total_sedentary_minutes / sum(activity$TotalMinutes)) * 100
lightly_active_percentage <- (total_lightly_active_minutes / sum(activity$TotalMinutes)) * 100
fairly_active_percentage <- (total_fairly_active_minutes / sum(activity$TotalMinutes)) * 100
very_active_percentage <- (total_very_active_minutes / sum(activity$TotalMinutes)) * 100

# Formatação
formatted_sedentary_percentage <- sprintf("%.2f%%", sedentary_percentage)
formatted_lightly_active_percentage <- sprintf("%.2f%%", lightly_active_percentage)
formatted_fairly_active_percentage <- sprintf("%.2f%%", fairly_active_percentage)
formatted_very_active_percentage <- sprintf("%.2f%%", very_active_percentage)

# Novo dataframe
formatted_percentage_data <- data.frame(
  Nivel_de_Atividade = c("Sedentário", "Levemente ativo", "Razoavelmente ativo", "Muito ativo"),
  Porcentagem = c(formatted_sedentary_percentage, formatted_lightly_active_percentage, 
                 formatted_fairly_active_percentage, formatted_very_active_percentage)
)

# Imprime
formatted_percentage_data
````

&nbsp; &nbsp; &nbsp; Podemos observar que a maioria dos usuários apresentam **81,3%** da sua atividade diária em minutos sedentários e apenas **1,74%** em minutos muito ativos.

Nota-se que *Terça e Sábado* são os dias com as médias mais altas de passos, sugerindo um maior engajamento em atividades físicas nesses dias. Por outro lado, *Domingo* apresenta a menor média, o que pode indicar um dia de descanso para a maioria dos usuários.

````r
merged_byhour$Hour <- factor(format(as.POSIXct(merged_byhour$Time, format="%I:%M:%S %p"), "%H"), 
                             levels = sprintf("%02d", 0:23))

steps_hour_avg <- merged_byhour %>%
  group_by(Hour) %>%
  summarize(AvgSteps = mean(StepTotal, na.rm = TRUE),
            AvgIntensity = mean(TotalIntensity, na.rm = TRUE))

ggplot(steps_hour_avg, aes(x = Hour, y = AvgSteps, fill = AvgIntensity)) +
  geom_bar(stat = "identity") +
  scale_fill_gradient(low = "lightblue", high = "pink") +
  labs(x = "Hora do Dia", y = "Média de Passos", title = "Média de Passos por Hora", fill="Intensidade") +
  theme_minimal() +
  theme(
    axis.text.x = element_text(angle = 90, hjust = 1, size = 10),
    axis.text.y = element_text(size = 10),
    plot.title = element_text(size = 20),
    plot.subtitle = element_text(size = 12)
  )
````

&nbsp; &nbsp; &nbsp; O gráfico '**Média de Passos por Hora**' destaca claramente os períodos do dia em que os usuários são mais ativos. Ele indica que a atividade tende a aumentar progressivamente pela manhã, com um pico de passos e intensidade nas horas centrais do dia. Após este pico, há uma diminuição gradual, refletindo padrões típicos de trabalho e descanso. Este insight pode ser usado para incentivar atividades nos períodos de menor movimento, como início da manhã e meio da tarde.

````r
heart_rate$Hour <- format(as.POSIXct(heart_rate$Time, format="%I:%M:%S %p"), "%H")

bpm_hour_avg <- heart_rate %>%
  group_by(Hour) %>%
  summarize(AvgBPM = mean(Value, na.rm = TRUE))

ggplot(bpm_hour_avg, aes(x = as.numeric(Hour), y = AvgBPM)) +
  geom_line(linewidth = 1.5, color = "#00BFC4") + 
  geom_point(color = "darkorange", size = 3) +  
  geom_smooth(se = FALSE, color = "darkgray", linetype = "dotted", size=2.5) + 
  scale_x_continuous(breaks = 0:23, labels = sprintf("%02d", 0:23)) +
  scale_y_continuous(labels = scales::comma) + 
  labs(
    x = "Hora do Dia",
    y = "Média de BPM",
    title = "Variação Média do Ritmo Cardíaco por Hora do Dia"
  ) +
  theme_minimal() +
  theme(
    axis.text.x = element_text(angle = 90, hjust = 1, size = 10),
    axis.text.y = element_text(size = 10),
    plot.title = element_text(size = 20),
    panel.grid.minor = element_blank()
  )
````

&nbsp; &nbsp; &nbsp; Já este gráfico mostra a **variação média do ritmo cardíaco por hora**, refletindo um padrão claro que acompanha os níveis de atividade física. Há um aumento notável do ritmo cardíaco durante as horas em que a atividade de passos *(do gráfico anterior)* é mais intensa, com picos nas horas centrais do dia. À medida que a atividade de passos diminui, especialmente no final da tarde e noite, o ritmo cardíaco também cai, indicando períodos de descanso. Essa sincronia entre o ritmo cardíaco e os passos diários realça a relação entre a atividade física e a resposta cardiovascular dos usuários.

````r
activity$DayOfWeek <- factor(activity$DayOfWeek, 
                             levels = c("Domingo", "Segunda-feira", "Terça-feira", 
                                        "Quarta-feira", "Quinta-feira", "Sexta-feira", "Sábado"))

calories_day_avg <- activity %>%
  group_by(DayOfWeek) %>%
  summarize(AvgCalories = mean(Calories, na.rm = TRUE))

ggplot(calories_day_avg, aes(x = DayOfWeek, y = AvgCalories, fill = DayOfWeek)) +
  geom_bar(stat = "identity") +
  geom_label(aes(label = scales::comma(round(AvgCalories))),
             vjust = -0.3, size = 3.8, color = "black", fill = "white", label.size = 0.4) +
  scale_fill_brewer(palette = "Set3") +
  labs(x = "Dia da Semana", y = "Média de Calorias Queimadas", title = "Média de Calorias Queimadas por Dia da Semana") +
  theme_minimal() +
  theme(
    axis.text.x = element_text(angle = 45, hjust = 1, size = 12),
    axis.text.y = element_text(size = 12),
    plot.title = element_text(size = 20),
    legend.position = "none"
  )
````

&nbsp; &nbsp; &nbsp; O gráfico de **média de calorias queimadas por dia da semana** revela que *Terça-feira* é o dia com o maior gasto calórico médio, enquanto *Quinta-feira* apresenta a menor média. Isso pode refletir um padrão semanal no comportamento dos usuários, onde o meio da semana motiva mais atividade física. A consistência no gasto calórico ao longo da semana sugere que os usuários mantêm um nível relativamente estável de atividade diária.

#### Incentivo à uma Vida Saudável Geral: 

&nbsp; &nbsp; &nbsp; Gerar lembretes de ingestão apropriada de água com notificações para os usuários em intervalos de tempo. Com os dados sobre as calorias queimadas, a Bellabeat pode oferecer conselhos nutricionais para ajudar os usuários a entender melhor como equilibrar a ingestão de calorias com o gasto energético.
  * Oferecer recursos úteis para ajudar os usuários na fase do sono – por exemplo: exercícios de respiração, podcasts com música relaxante, técnicas de sono.

&nbsp; &nbsp; &nbsp;
#### Foco em Horários de Pico de Atividade: 

&nbsp; &nbsp; &nbsp; Utilizar as horas do dia com maior atividade física para enviar lembretes motivacionais ou sugerir curtos exercícios que podem ajudar a aumentar ainda mais o nível de atividade dos usuários. Outros tipos de notificações:
  * Se a frequência cardíaca em repouso do usuário variar significativamente do normal.
  * Incentivar atividades caso um usuário tenha passado uma hora acordado na cama (fora do horário de sono).
  * Incentivar a atividade se um usuário estiver sedentário por um longo período de tempo.

&nbsp; &nbsp; &nbsp;
#### Promover Estilos de Vida +Ativos:
&nbsp; &nbsp; &nbsp; Usar as informações sobre a distribuição dos níveis de atividade diária para promover estilos de vida ativos, focando em converter minutos sedentários em levemente ativos e aumentar a porcentagem de atividades mais vigorosas. Aqui também pode-se enviar lembretes motivacionais e sugerir curtos exercícios.

&nbsp; &nbsp; &nbsp;
#### Sistema de Conquistas e Criação de Comunidade: 
&nbsp; &nbsp; &nbsp; Criar desafios e recursos de compartilhamento social no aplicativo para aumentar o engajamento dos usuários e incentivá-los a manter ou melhorar seus hábitos saudáveis.
  * Notificar o sucesso e manter os desafios semanais de condicionamento físico também devem ser integrados no aplicativo. Essas notificações ou alertas criam reforços positivos que estimularão os usuários a realizar mais. 
  * Exemplo: cumprir os 10.000 passos recomendados e pelo menos 7 horas de sono diárias.
  * Pode ser interessante também combinar com um sistema de incentivo de premiação de pontos em que os usuários receberão pontos Bellabeat em produtos/assinaturas.

&nbsp; &nbsp; &nbsp;
#### Ciclo Menstrual e Personalização do Treino:
&nbsp; &nbsp; &nbsp; As mulheres têm necessidades fisiológicas específicas e padrões de atividade que podem variar significativamente ao longo do ciclo menstrual. Oferecer personalização com base nas fases do ciclo pode ajudar a melhorar a eficácia dos exercícios e o bem-estar geral.
  * Incentivar atividades mais intensas durante a fase folicular e recomendar exercícios mais leves ou de recuperação durante a fase lútea. 
  * Fornecer informações educacionais sobre como o ciclo menstrual pode afetar a atividade física, o sono e o metabolismo. Isso pode incluir dicas sobre nutrição e hidratação adequadas em diferentes fases do ciclo.

&nbsp; &nbsp; &nbsp;
#### Uso constante: 
&nbsp; &nbsp; &nbsp; Promover características de qualidade nos produtos Bellabeat para que os usuários utilizem os dispositivos o tempo todo durante todo o dia. Exemplos: bateria de longa duração, resistência a água, produtos confortáveis e modernos.

&nbsp; &nbsp; &nbsp;
#### Feedback Personalizado: 
&nbsp; &nbsp; &nbsp; Desenvolver um sistema de feedback contínuo no aplicativo que forneça aos usuários atualizações regulares sobre seu progresso e sugestões personalizadas baseadas em seus dados de atividade e saúde.
